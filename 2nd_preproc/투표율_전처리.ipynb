```python
import pandas as pd
```

```python
from openpyxl import load_workbook

def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

```python
def preproc(df, arg):
    df = df.reset_index(names=["loc_2", "arg_1", "arg_2"])
    df = df[(df["loc_2"] != "합계") & (df["arg_1"] != "합계") & (df["arg_2"] != "선거인수")].copy()
    df["loc_1"] = arg

    df["18_29"] = df.loc[:, ["18세", "19세", "20-24세", "25-29세"]].sum(axis=1)
    df["30_39"] = df.loc[:, ["30-34세", "35-39세"]].sum(axis=1)
    df["40_49"] = df.loc[:, ["40-49세"]].sum(axis=1)
    df["50_59"] = df.loc[:, ["50-59세"]].sum(axis=1)
    df["60_"] = df.loc[:, ["60-69세", "70-79세", "80세이상"]].sum(axis=1)
    df = df.loc[:, ["loc_1", "loc_2", "arg_1", "18_29", "30_39", "40_49", "50_59", "60_"]]
    return df


mapping = {
    "강원": "강원도",
    "경기": "경기도",
    "경남": "경상남도",
    "경북": "경상북도",
    "광주": "광주광역시",
    "대구": "대구광역시",
    "대전": "대전광역시",
    "부산": "부산광역시",
    "서울": "서울특별시",
    "세종": "세종특별자치시",
    "울산": "울산광역시",
    "인천": "인천광역시",
    "전남": "전라남도",
    "전북": "전라북도",
    "제주": "제주특별자치도",
    "충남": "충청남도",
    "충북": "충청북도",
}
```

```python
sheetnames = get_sheetnames_xlsx(
    "./../raw_data/제_8_전국동시지방선거_투표율/01_투표율분석결과(표본)/12.선거인의 성별·연령대별 분포(구시군별).xlsx"
)
sheetnames
```

```python
preproc_df_list = list()
for sheetname in sheetnames:
    raw_df = pd.read_excel(
        "./../raw_data/제_8_전국동시지방선거_투표율/01_투표율분석결과(표본)/12.선거인의 성별·연령대별 분포(구시군별).xlsx",
        sheet_name=sheetname,
        header=[3],
        index_col=[0, 1, 2],
    )
    df = raw_df.copy()
    preproc_df = preproc(df, sheetname)
    preproc_df_list.append(preproc_df)
전국동시지방선거_8_df = pd.concat(preproc_df_list, axis=0)
전국동시지방선거_8_df["loc_1"] = 전국동시지방선거_8_df["loc_1"].map(mapping)
전국동시지방선거_8_df
```

```python
전국동시지방선거_8_df.to_csv("./2nd_preproc_data/전국동시지방선거_투표율_8_df.csv")
```
